```python
import pandas as pd
import numpy as np

trainDB = pd.read_csv('train.csv')

trainDB.head()
```

```python
trainDB['MonthYear'] = trainDB['MoSold'].map(str) + '-' + trainDB['YrSold'].map(str)
    
trainDB.head()
```

```python
trainDB.describe()
```

```python
import matplotlib.pyplot as plt
import datetime as dt

def rolling_mean(a, w):
    r = np.empty(len(a))
    r.fill(0)  
    for i in range(w, len(a), 1):
        r[i] = np.mean(a[i-w:i])
    return r

months = []
bins = []

for y in range(2006, 2011,1):
    for m in range(1,13,1):
        thisMonth = str(m) + '-' + str(y)
        months.append(dt.datetime(y, m, 1, 0, 0))
        bins.append(trainDB[trainDB["MonthYear"].apply(str) == thisMonth].Id.count())

movingAvg = rolling_mean(bins, 3)
        
plt.figure(figsize=(10,6))
plt.bar(months, bins, 20, label='# sales')
plt.plot(months, movingAvg, 'r--', label='3-mo moving avg')
plt.legend(loc='upper right')
plt.xlabel('Month of Sale')
plt.ylabel('# of Sales')
plt.show()
```

```python
soldByMonth = np.empty(12)

for m in range(1,13,1):
        monthlyCount = trainDB[trainDB["MoSold"] == m].Id.count()
        if not np.isnan(monthlyCount):
            soldByMonth[m-1] = monthlyCount

plt.figure(figsize=(10,4))
plt.bar(range(1, 13, 1), soldByMonth, .5, label='# of sales', color='green', alpha=0.4)
plt.xticks(range(1, 13))
plt.grid(True)
plt.xlabel('Month of Sale')
plt.ylabel('# of sales')
plt.show()
```

```python
avgByMonth = np.empty(12)
sdevByMonth = np.empty(12)

for m in range(1,13,1):
        monthlyAvg = trainDB[trainDB["MoSold"] == m].SalePrice.mean()
        monthlySDev = trainDB[trainDB["MoSold"] == m].SalePrice.std()
        if not np.isnan(monthlyAvg):
            avgByMonth[m-1] = monthlyAvg
            sdevByMonth[m-1] = monthlySDev

plt.figure(figsize=(10,4))
plt.errorbar(range(1,13), avgByMonth, sdevByMonth, linestyle='none', marker='^')
plt.xticks(range(1, 13))
plt.grid(True)
plt.xlabel('Month of Sale')
plt.ylabel('Average Sale Price')
plt.show()
```

Useless!

### Sales over time

```python
avgPriceSoldByMonth = np.empty(12*5)
medianPriceSoldByMonth = np.empty(12*5)

for y in range(2006, 2011,1):
    for m in range(1,13,1):
        thisMonth = str(m) + '-' + str(y)
        monthlyMean = trainDB[trainDB["MonthYear"].apply(str) == thisMonth].SalePrice.mean()
        monthlyMedian = trainDB[trainDB["MonthYear"].apply(str) == thisMonth].SalePrice.median()
        if not np.isnan(monthlyMean):
            avgPriceSoldByMonth[(y-2006)*12+(m-1)] = monthlyMean
            medianPriceSoldByMonth[(y-2006)*12+(m-1)] = monthlyMedian

movingAvg = rolling_mean(avgPriceSoldByMonth, 3)

fig = plt.figure(figsize=(10,6))
plt.bar(months, avgPriceSoldByMonth, 20, label='average asking price', color='green', alpha=0.4)
plt.plot(months, movingAvg, 'r--', label='3-mo moving avg')
plt.plot(months, medianPriceSoldByMonth, 'g*', label='median asking price')
plt.legend(loc='upper right')
plt.xlabel('Month of Sale')
plt.ylabel('Avg sale price')
plt.show()
```

```python
normedMedian = np.empty(12*5)
for y in range(2006, 2011,1):
    for m in range(1,13,1):
        i = (y-2006)*12+(m-1)
        normedMedian[i] = medianPriceSoldByMonth[i]/medianPriceSoldByMonth[0]

plt.figure(figsize=(10,4))
plt.plot(months[:12*4+6], normedMedian[:12*4+6], 'g', label='median sale price')
plt.ylabel('Median Sale Price \n(normalised to Jan \'06)')
plt.xlabel('Month of Sale')
plt.grid(True)
plt.show()
```

Is this the effect of the 2008 financial crisis?
The median sale price peaks in Dec '07 and then tends to stay below its Jan '06 value.

## House Quality

### Overall Quality

```python
soldByQual = np.empty(10)

for q in range(1,11,1):
    qualCount = trainDB[trainDB["OverallQual"] == q].Id.count()
    if not np.isnan(qualCount):
        soldByQual[q-1] = qualCount

plt.figure(figsize=(10,6))
plt.bar(range(1, 11, 1), soldByQual, .5, label='# of sales', color='lightblue', alpha=0.8)
plt.xticks(range(1, 11))
plt.grid(True)
plt.xlabel('Overall Quality')
plt.ylabel('# of sales')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(trainDB['OverallQual'], trainDB['SalePrice'])
plt.grid(True)
plt.yticks(range(0,800001,50000))
plt.show()
```

```python
avgPriceSoldByQual = np.empty(10)
medianPriceSoldByQual = np.empty(10)
sdevPriceSoldByQual = np.empty(10)

for q in range(1,11):
    qualMean = trainDB[trainDB["OverallQual"] == q].SalePrice.mean()
    qualMedian = trainDB[trainDB["OverallQual"] == q].SalePrice.median()
    qualSDev = trainDB[trainDB["OverallQual"] == q].SalePrice.std()
    if not np.isnan(qualMean):
        avgPriceSoldByQual[q-1] = qualMean
        medianPriceSoldByQual[q-1] = qualMedian
        sdevPriceSoldByQual[q-1] = qualSDev

fig = plt.figure(figsize=(10,6))
plt.bar(range(1,11), avgPriceSoldByQual, .5, label='average asking price', color='orange', alpha=0.4)
plt.plot(range(1,11), medianPriceSoldByQual, marker='o', color='brown', linestyle='none', label='median asking price')
plt.errorbar(range(1,11), avgPriceSoldByQual, sdevPriceSoldByQual, linestyle='none', marker='', color='orange', label='standard deviation')
plt.xticks(range(1,11))
plt.yticks(range(0,800001,50000))
plt.grid(True)
plt.legend(loc='upper left')
plt.xlabel('Overall Quality')
plt.ylabel('Sale Price')
plt.show()
```
